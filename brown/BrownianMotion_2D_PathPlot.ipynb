```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(100)

    
T = 1
N = 501
dt = T/(N-1)
t = np.linspace(dt,T,N)

dX = np.sqrt(dt) * np.random.randn(1,N)
X = np.cumsum(dX, axis=1)
    
dY = np.sqrt(dt) * np.random.randn(1,N)
Y = np.cumsum(dY, axis=1)
    

fig, ax = plt.subplots()
ax.plot(X[0,:], Y[0,:], color='orchid')
ax.plot(X[0,0], Y[0,0], 'bo')
ax.plot(X[0,-1], Y[0,-1], 'go')
ax.set_xlabel('$X(t)$', fontsize=14)
plt.ylabel('$Y(t)$', fontsize=14, rotation=0, labelpad=20)
plt.tight_layout()
plt.show()

# Uncomment to save the image
#fig.savefig('brownian_2d.png', dpi=600)
```

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt

np.random.seed(10)

fig = plt.figure()

T = 1
N = 501 # Number of points, number of subintervals = N-1
dt = T/(N-1) # Time step
t = np.linspace(0,T,N)

# Preallocate arrays for efficiency:
dX = [0]*N 
X = [0]*N

# Initialization:
dX[0] = np.sqrt(dt)*np.random.randn() # Eq. (3)
X[0] = dX[0]

for i in range(1,N):
    dX[i] = np.sqrt(dt)*np.random.randn() # Eq. (3)
    X[i] = X[i-1] + dX[i] # Eq. (4)


plt.plot(t, X, color='orchid')
plt.xlabel('t', fontsize=14)
plt.ylabel('$X(t)$', fontsize=14, rotation=0, labelpad=20)
axes = plt.gca()
axes.set_xlim([0,T])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()

fig.savefig('brownian_1d.pdf', dpi=600)
```

```python
np.random.seed(10)

fig = plt.figure()
        
T = 1
N = 501 # Number of points, number of subintervals = N-1
dt = T/(N-1) # Time step
t = np.linspace(0,T,N)
M = 20 # Number of walkers/paths

dX = np.sqrt(dt) * np.random.randn(M, N)
X = np.cumsum(dX, axis=1)


for i in range(M):
    plt.plot(t, X[i,:])
  
    
plt.xlabel('$t$', fontsize=14)
plt.ylabel('$X(t)$', fontsize=14, rotation=0, labelpad=10)
axes = plt.gca()
axes.set_xlim([0,1])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()

fig.savefig('brownian_1d_multi.png', dpi=600)
```

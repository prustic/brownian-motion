```python
import numpy as np
from scipy.stats import cauchy
from matplotlib import pyplot as plt
gamma_values = [0.5, 1.0, 2.0]
linestyles = ['-', '--', ':']
mu = 0
x = np.linspace(-10, 10, 1000)

fig, ax = plt.subplots(figsize=(12,6))

for gamma, ls in zip(gamma_values, linestyles):
    dist = cauchy(mu, gamma)

    plt.plot(x, dist.pdf(x), ls=ls, color='black',
             label=r'$\mu=%i,\ \gamma=%.1f$' % (mu, gamma))

plt.xlim(-4.5, 4.5)
plt.ylim(0, 0.65)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\mu,\gamma)$', rotation=0, labelpad=25)

plt.legend()
plt.show()

fig.savefig('cauchy.pdf', dpi=600)
```

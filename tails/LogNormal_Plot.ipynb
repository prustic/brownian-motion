```python
from scipy.stats import lognorm
```

```python
import numpy as np
import pylab as plt

x=np.linspace(0,6,200)

fig = plt.figure()

stddev = 0.25
mean = 0
dist=lognorm([stddev],loc=mean)
plt.plot(x,dist.pdf(x), label='μ = 0, σ = 0.25')


stddev = 0.5
mean = 0
dist=lognorm([stddev],loc=mean)
plt.plot(x,dist.pdf(x), label='μ = 0, σ = 0.5')

stddev = 1
mean = 0
dist=lognorm([stddev],loc=mean)
plt.plot(x,dist.pdf(x), label='μ = 0, σ = 1')

plt.grid(b=True, color='grey', alpha=0.3, linestyle='-.', linewidth=2)
plt.xlabel('x', fontsize=14)
plt.ylabel('Pr(X=x)', fontsize=14, rotation=0, labelpad=45)
plt.rcParams["figure.figsize"] = [12, 6]
plt.legend(loc='best')
plt.show()

fig.savefig('lognorm.pdf', dpi=600)
```

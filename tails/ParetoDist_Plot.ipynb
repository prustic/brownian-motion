```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

fig = plt.figure()

x_m = 1 #scale
alpha = [1, 2, 3] #list of values of shape parameters
samples = np.linspace(start=0, stop=5, num=1000)
for a in alpha:
    output = np.array([pareto.pdf(x=samples, b=a, loc=0, scale=x_m)])
    plt.plot(samples, output.T, label='Î± = {0}' .format(a))
plt.xlabel('x', fontsize=14)
plt.ylabel('Pr(X=x)', fontsize=14, rotation=0, labelpad=45)
plt.grid(b=True, color='grey', alpha=0.3, linestyle='-.', linewidth=2)
plt.rcParams["figure.figsize"] = [12, 6]
plt.legend(loc='best')
plt.show()

fig.savefig('pareto.pdf', dpi=600)
```
